normalization 消除數據間的差異, 且可計算的範圍變大.
Make sure features are on a similar scale.
標準化數據  preprocessing.scale

```python
from sklearn import preprocessing
import numpy as np

data_features=np.array([[10,2.7,3.6],
                        [-100,5,-2],
                        [120,20,40]],dtype=np.float64)
```

```python
#未標準化的data_features
print(data_features)
```

```python
#標準化後的data_features  => scale會變成 -1.X ~ 1.X   縮短之間的差異性
print(preprocessing.scale(data_features))   
```

```python
from sklearn.model_selection import train_test_split  #將train data 及test data分開的模塊
from sklearn.datasets.samples_generator import make_classification #分類
from sklearn.svm import SVC   #SVM分類演算法 的 support vectors classification
import matplotlib.pyplot as plt

#生成data     n_samples=樣本數,n_features=屬性數,n_redundant=0,n_informative=相關屬性數,random_state=隨機產生,n_clusters_per_class=1,scale=100
X,y=make_classification(n_samples=300,n_features=2,n_redundant=0,n_informative=2,random_state=22,n_clusters_per_class=1,scale=100)
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()     #查看數據結構
```

```python
#未標準化的數據
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.3)  #Train去學習, test去預測
classification_=SVC()
classification_.fit(X_train,y_train)
print(classification_.score(X_test,y_test))
```

```python
#標準化scale 數據
# X=preprocessing.minmax_scale(X,feature_range=(-1,1)) #將數據標準化到 -1 ~ 1的範圍
X=preprocessing.scale(X)  ####
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.3)  #Train去學習, test去預測
classification_=SVC()
classification_.fit(X_train,y_train)
print(classification_.score(X_test,y_test))
```
